```python
pib_15_19 = pd.read_excel('raw_data/pib_muni_2019-2015.xls')

#Cleaning excel table into dataframe

pib_15_19.drop([0,1,2,3,4,5,6], axis=0, inplace=True) #Excel headers out
pib_1519 = pib_15_19.rename(columns=pib_15_19.iloc[0]) #Assign column headers
pib_1519.columns.values[0] = 'Municipio' #Column names
pib_1519.columns.values[1] = '2015'
pib_1519.columns.values[2] = '2016'
pib_1519.columns.values[3] = '2017p' #p stand for 'preliminar'
pib_1519.columns.values[4] = '2018a'
pib_1519.columns.values[5] = '2019a'
pib_1519.drop([7,8,10,22,202,203,204,205,206], axis=0, inplace=True) #More excel format out
pib_1519.reset_index(drop=True, inplace=True) #Reset index
pib_1519['2015'] = pd.to_numeric(pib_1519['2015'], downcast="integer") #small nums make lazy machines happy
pib_1519['2016'] = pd.to_numeric(pib_1519['2016'], downcast="integer")
pib_1519['2017p'] = pd.to_numeric(pib_1519['2017p'], downcast="integer")
pib_1519['2018a'] = pd.to_numeric(pib_1519['2018a'], downcast="integer")
pib_1519['2019a'] = pd.to_numeric(pib_1519['2018a'], downcast="integer")

pib_1519.head(15)
```
